```
import numpy as np
import pandas as pd
import sklearn

import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Conv2D, Convolution2D, MaxPooling2D, Flatten
from keras.optimizers import SGD, Adam, RMSprop, Adadelta
from keras.utils import np_utils
from keras import backend as K

import os

os.environ["TF_CPP_MIN_LOG_LEVEL"] = "2"

import matplotlib.pyplot as plt
%matplotlib inline
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('retina')


```

```
LEN = 1000
xs = np.linspace(-1,1,LEN, dtype=np.float32)


df = pd.DataFrame(data=np.array([xs]).T)

plt.rcParams['figure.dpi'] = 100
plt.ion()
plt.axis('equal')

df.columns = ['X']
radius1 = 0.3
radius2 = 0.6

df['Y_Circle1_Pos'] = np.sqrt((radius1**2) - df['X']**2)
df['Y_Circle2_Pos'] = np.sqrt((radius2**2)- df['X']**2)
df['Y_Circle1_Neg'] = -np.sqrt((radius1**2) - df['X']**2)
df['Y_Circle2_Neg'] = -np.sqrt((radius2**2)- df['X']**2)



#print(df_ones.head())

#plt.imshow(im, cmap='gray')
plt.scatter(df['X'], df['Y_Circle1_Pos'], c='pink')
plt.scatter(df['X'], df['Y_Circle2_Pos'], c='purple')
plt.scatter(df['X'], df['Y_Circle1_Neg'], c='pink')
plt.scatter(df['X'], df['Y_Circle2_Neg'], c='purple')


#plt.savefig('digit_img.png')
```
