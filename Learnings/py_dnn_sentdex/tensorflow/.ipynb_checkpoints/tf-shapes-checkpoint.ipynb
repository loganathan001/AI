```python
import tensorflow as tf
import numpy as np

tf.enable_eager_execution()

t = tf.constant([3,2,4,5,6,7])

print(t.shape)
print(t)
t0 = tf.reshape(t, [3,2])
print(t0.shape)
print(t0)
t1 = tf.expand_dims(t0, 1)
print(t1.shape)
print(t1)
t11 = tf.expand_dims(t0, 2)
print(t11.shape)
print(t11)
t12 = tf.expand_dims(t0, 0)
print(t12.shape)
print(t12)
t2 = tf.slice(t0, [0, 1], [2, 1])
print(t2.shape)
print(t2)
t3 = t0[:2, 1]
print(t3.shape)
print(t3)

t4 = tf.squeeze(t1)
print(t4.shape)
print(t4)

t5 = tf.squeeze(t11)
print(t5.shape)
print(t5)

t6 = tf.squeeze(t12)
print(t6.shape)
print(t6)
```
